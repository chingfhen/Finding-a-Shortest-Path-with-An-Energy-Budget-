```python
def heuristic_sl_distance(node1, node2):
    x_dist = abs(Coord[node1][0])-abs(Coord[node2][0])
    y_dist = abs(Coord[node1][1])-abs(Coord[node2][1])
    return math.sqrt(x_dist**2 + y_dist**2)
```
