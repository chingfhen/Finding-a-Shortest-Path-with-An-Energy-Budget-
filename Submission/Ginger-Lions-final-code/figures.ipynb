```python
# plot times for different alpha values
data = []

for alpha in range(10):

    # average time for 20 runs heuristic A*
    times = []
    for i in range(20):
        start = time.time()
        shortest = find_path_astar(G, "1", "50", cost_func=distance_func, energy_func=energy_func,
                                heuristic_func=heuristic_sl_distance, alpha=alpha, energy_budget=287932)
        end = time.time()
        times.append(end-start)

    avg_time = sum(times)/len(times)
    distance = shortest[1]
    data.append((avg_time, distance))
```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame(data, columns=['time', 'distance'])
fig = sns.lineplot(data=df, x=range(1, len(df)+1), y='time')
fig.set_title("Avg Runtime vs Alpha Value")
fig.set_xlabel("alpha")
fig.set_ylabel("runtime (seconds)")
```

```python
fig = sns.lineplot(data=df, x=range(1, len(df)+1), y='distance')
fig.set_title("Path Distance vs Alpha Value")
fig.set_xlabel("alpha")
fig.set_ylabel("path distance (m)")
```
